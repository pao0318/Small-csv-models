```python
df_train['Dependents'] = df_train['Dependents'].replace(['3+'],'3')
df_test['Dependents'] = df_test['Dependents'].replace(['3+'],'3')
```

```python
df_test = df_test.fillna(value = 1, axis = 0)
df_train = df_train.fillna(value = 1, axis = 0)
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python

from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
df_train['Qual_var']= le.fit_transform(df_train['Qual_var']) 
df_test['Qual_var']= le.fit_transform(df_test['Qual_var'])#Grad=0,Nongrad=1
```

```python
df_train['Prop_Area']= le.fit_transform(df_train['Prop_Area'])
df_test['Prop_Area']= le.fit_transform(df_test['Prop_Area'])
```

```python
df_train['Sex'] = df_train['Sex'].str.replace('M','1')
df_train['Sex'] = df_train['Sex'].str.replace('F','0')

df_train['SE'] = df_train['SE'].str.replace('N','1')
df_train['SE'] = df_train['SE'].str.replace('Y','0')


df_train['Marital_Status'] = df_train['Marital_Status'].str.replace('N','1')
df_train['Marital_Status'] = df_train['Marital_Status'].str.replace('Y','0')


```

```python
df_test['Sex'] = df_test['Sex'].str.replace('M','1')
df_test['Sex'] = df_test['Sex'].str.replace('F','0')

df_test['SE'] = df_test['SE'].str.replace('N','1')
df_test['SE'] = df_test['SE'].str.replace('Y','0')


df_test['Marital_Status'] = df_test['Marital_Status'].str.replace('N','1')
df_test['Marital_Status'] = df_test['Marital_Status'].str.replace('Y','0')
```

```python
df_test = df_test.fillna(value = 1, axis = 0)
df_train = df_train.fillna(value = 1, axis = 0)
```

```python
df_train.head(15)
```

```python
df_test.head(15)
```

```python
X = df_train.drop('CPL_Status', axis = 1)
y = df_train['CPL_Status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=35)
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train,y_train)

from sklearn.svm import SVC
model2 = SVC()
model2.fit(X_train,y_train)

from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors = 5)
model3.fit(X_train,y_train)

from sklearn.tree import DecisionTreeClassifier
model4 = DecisionTreeClassifier()
model4.fit(X_train,y_train)

from sklearn.ensemble import RandomForestClassifier
model5 = RandomForestClassifier()
model5.fit(X_train,y_train)

from xgboost import XGBClassifier
model6 = XGBClassifier()
model6.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
prediction1 = model1.predict(X_test)
prediction2 = model2.predict(X_test)
prediction3 = model3.predict(X_test)
prediction4 = model4.predict(X_test)
prediction5 = model5.predict(X_test)
prediction6 = model6.predict(X_test)
```

```python
print(classification_report(y_test,prediction1))
```

```python
print(classification_report(y_test,prediction2))
```

```python
print(classification_report(y_test,prediction3))
```
